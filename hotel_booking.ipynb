## ***DATA CLEANING***

***Drop Unnecessary Data***

```python
df = df.drop(['name', 'email', 'phone-number', 'credit_card'], axis=1)
```

```python
df.drop(['agent', 'company'], axis=1, inplace=True)
```

```python
df = df[df['adr']<2000]
```

***Change Data Types***

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

## ***DATA ANALYST & VISUALIZATION***

```python
#Check the cancelled percentage

cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
plt.figure(figsize = (5,4))
plt.title("Reservation Status Count")
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
# viz the cancelled percentage based on hotel types

plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotel types', size = 20)
plt.xlabel('Hotel')
plt.ylabel('No. of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

***Hypothesis : High price making customers cancel their booking***

```python
# Check if price has any effect on hotel cancellations, so get average daily rate(adr) for each day

resort_hotel = df[df['hotel'] == 'Resort Hotel']
city_hotel = df[df['hotel'] == 'City Hotel']

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Create visualization for checking the hypothesis

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City & Resort Hotels', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
# Find out the month with maximum reservations & cancellations
# Create new column for month

df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'dark')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('Month')
plt.ylabel('No. of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
# Check the price for each month for cancelled bookings data

plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.legend(fontsize=20)
plt.show()
```

***In this poin, we can see that cancellations are more likely occurs when the price is high.***

```python
# Check cancellation rates per country
# As the no. of countries is 177, so we will only check the top 10 countries

cancelled_data = df[df['is_canceled'] == 1]  # Data frame for cancelled reservations
top_10_countries = cancelled_data['country'].value_counts()[:10]  # Countries which are the top 10 in count

# Create pie chart to show the data for these 10 countries

plt.figure(figsize = (8,8))
plt.title('Top 10 countries with cancelled reservations')
plt.pie(top_10_countries, autopct = '%.1f', labels = top_10_countries.index)  # Pie chart with data upto 1 decimal places
plt.show()
```

***Portugal (or PRT) has the most cancellations, which is about 70.1 %***

***Hypothesis 2 : The hotel company needs to improve their marketing, run some campaign, do the promotion, specifically in Portugal by Online media***

```python
# Check the market segment

df['market_segment'].value_counts()
```

```python
# Get the percentage form of the above data

df['market_segment'].value_counts(normalize = True)
```

This poin, we can see that ONLINE booking has 47% and only 20% from the OFFLINE agent. 

```python
# Check the same count for cancelled reservation data

cancelled_data['market_segment'].value_counts(normalize = True)
```

Most of the cancellations are also being done by customers coming through ONLINE booking by 46%

```python
# Check if the ADR is high for the cancelled reservation or non-cancelled reservations
# ADR for cancelled reservations

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

# Create data frame for non-cancelled reservations

not_cancelled_data = df[df['is_canceled'] == 0]

# ADR for non-cancelled reservations

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

# Plot the data

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled')
plt.legend()
```

```python
# Filter out data as the above data has inconsistencies before 2016 due to having less data for that time period

cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016')]
```

```python
# Plot the filtered data
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled')
plt.legend(fontsize = 20)
plt.show()
```

***Here is the another proof that high rates (price) is important factor that contributing in cancellations rates.***
